```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/dataset_real (1).csv")
df.head()
```

```python
df.corr()
```

```python
import statsmodels.api as sm
inputs = ['google_search_spent','google_performance_max_spent','fb_retargeting_spent','fb_prospecting_spent','google_organico','Sconti']
x = df[inputs]
y = df['revenue']
x = sm.add_constant(x)

result = sm.OLS(y,x).fit()

print(result.summary())
```

```python
inputs = ['google_search_spent','Sconti']
x = df[inputs]
y = df['revenue']
x = sm.add_constant(x)

result = sm.OLS(y,x).fit()

print(result.summary())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x,y, shuffle = False, test_size = 0.5 )
model = LinearRegression()
model.fit(X_train, Y_train)
print(
    model.score(X_train, Y_train),
    model.score(X_test, Y_test)
)
```
