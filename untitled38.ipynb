```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/uber-raw-data-sep14.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
column_name = df.columns.tolist()

for i in range (len(column_name)):
  df[column_name] = df[column_name].fillna(0)

df.isnull().sum()
```

```python
df["Date/Time"] = df["Date/Time"].map(pd.to_datetime)
df.head()
```

```python
df["Day"] = df["Date/Time"].apply(lambda x: x.day)
df["Weekdays"] = df["Date/Time"].apply(lambda x: x.weekday())
df["Hour"]  = df["Date/Time"].apply(lambda x: x.hour)
df.head()
```

```python
import seaborn as sns
sns.set(rc = {'figure.figsize':(12,10)})
sns.displot(df['Day'])
```

```python
sns.displot(df['Weekdays'])
```

```python
sns.displot(df['Hour'])
```

```python
df1 = df.groupby(["Weekdays","Hour"]).apply(lambda x:len(x))
df1 = df1.unstack()
sns.heatmap(df1,annot=False)
```

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='Lon', y='Lat', alpha=0.4, s=df['Day'], label='Uber Trips',
figsize=(12, 8), cmap=plt.get_cmap('jet'))
plt.title("Uber Trips Analysis")
plt.legend()
plt.show()
```
